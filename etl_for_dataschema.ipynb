```python
# Columns to melt into long format
years = ['2018', '2019', '2020', '2021', '2022']
metrics = ['Tot_Spndng', 'Tot_Dsg_Unts', 'Tot_Clms', 'Tot_Benes', 
           'Avg_Spnd_Per_Dsg_Unt_Wghtd', 'Avg_Spnd_Per_Clm', 
           'Avg_Spnd_Per_Bene', 'Outlier_Flag']
```

```python
metrics2 = ['Tot_Spndng', 'Tot_Dsg_Unts', 'Tot_Clms', 
           'Avg_Spnd_Per_Dsg_Unt_Wghtd', 'Avg_Spnd_Per_Clm'
           , 'Outlier_Flag']
```

```python
# Create a list of tuples for melting
melt_vars = [(f'{metric}_{year}', metric, year) for metric in metrics for year in years]
```

```python
melt_vars2 = [(f'{metric}_{year}', metric, year) for metric in metrics2 for year in years]
```

```python
melt_vars2
```

```python
# Melt data into long format
long_data = pd.DataFrame()
```

```python
long_data2= pd.DataFrame()
```

```python
for var, metric, year in melt_vars:
    temp_df = medicare_data[['Brnd_Name', 'Mftr_Name', var]].copy()
    temp_df.columns = ['Brnd_Name', 'Mftr_Name', 'value']
    temp_df['Year'] = int(year)
    temp_df['Metric'] = metric
    long_data = pd.concat([long_data, temp_df])
```

```python
medicaid_data.columns
```

```python
for var, metric, year in melt_vars2:
    temp_df2 = medicaid_data[['Brnd_Name', 'Mftr_Name', var]].copy()
    temp_df2.columns = ['Brnd_Name', 'Mftr_Name', 'value']
    temp_df2['Year'] = int(year)
    temp_df2['Metric'] = metric
    long_data2 = pd.concat([long_data2, temp_df2])
```

```python
# Pivot to get each metric in a column again
long_data = long_data.pivot_table(index=['Brnd_Name', 'Mftr_Name', 'Year'], 
                                  columns='Metric', 
                                  values='value').reset_index()
```

```python
# Pivot to get each metric in a column again
long_data2= long_data2.pivot_table(index=['Brnd_Name', 'Mftr_Name', 'Year'], 
                                  columns='Metric', 
                                  values='value').reset_index()
```

```python
# Reorder columns as specified
ordered_columns = [
    'Brnd_Name',
    'Mftr_Name',
    'Year',
    'Tot_Spndng',
    'Tot_Dsg_Unts',
    'Tot_Clms',
    'Tot_Benes',
    'Avg_Spnd_Per_Dsg_Unt_Wghtd',
    'Avg_Spnd_Per_Clm',
    'Avg_Spnd_Per_Bene',
    'Outlier_Flag'
]
```

```python
# Reorder columns as specified
ordered_columns2 = [
    'Brnd_Name',
    'Mftr_Name',
    'Year',
    'Tot_Spndng',
    'Tot_Dsg_Unts',
    'Tot_Clms',
    'Avg_Spnd_Per_Dsg_Unt_Wghtd',
    'Avg_Spnd_Per_Clm',
    'Outlier_Flag'
]
```

```python
#  Reorder the DataFrame columns
long_data2 = long_data2[ordered_columns2]

# Convert columns to the specified data types
long_data2['Brnd_Name'] = long_data2['Brnd_Name'].astype(str)
long_data2['Mftr_Name'] = long_data2['Mftr_Name'].astype(str)
long_data2['Year'] = long_data2['Year'].astype(np.int16)  # SMALLINT
long_data2['Tot_Spndng'] = long_data2['Tot_Spndng'].astype(np.float64)  # DECIMAL(15,2)
long_data2['Tot_Dsg_Unts'] = long_data2['Tot_Dsg_Unts'].astype(np.float64)  # DECIMAL(15,2)
long_data2['Tot_Clms'] = long_data2['Tot_Clms'].astype(np.int64)  # INT
long_data2['Avg_Spnd_Per_Dsg_Unt_Wghtd'] = long_data2['Avg_Spnd_Per_Dsg_Unt_Wghtd'].astype(np.float64)  # DECIMAL(15,2)
long_data2['Avg_Spnd_Per_Clm'] = long_data2['Avg_Spnd_Per_Clm'].astype(np.float64)  # DECIMAL(15,2)
long_data2['Outlier_Flag'] = long_data2['Outlier_Flag'].astype(bool)  # BOOLEAN
```

```python
#  Reorder the DataFrame columns
long_data = long_data[ordered_columns]

# Convert columns to the specified data types
long_data['Brnd_Name'] = long_data['Brnd_Name'].astype(str)
long_data['Mftr_Name'] = long_data['Mftr_Name'].astype(str)
long_data['Year'] = long_data['Year'].astype(np.int16)  # SMALLINT
long_data['Tot_Spndng'] = long_data['Tot_Spndng'].astype(np.float64)  # DECIMAL(15,2)
long_data['Tot_Dsg_Unts'] = long_data['Tot_Dsg_Unts'].astype(np.float64)  # DECIMAL(15,2)
long_data['Tot_Clms'] = long_data['Tot_Clms'].astype(np.int64)  # INT
long_data['Tot_Benes'] = long_data['Tot_Benes'].astype(np.int64)  # INT
long_data['Avg_Spnd_Per_Dsg_Unt_Wghtd'] = long_data['Avg_Spnd_Per_Dsg_Unt_Wghtd'].astype(np.float64)  # DECIMAL(15,2)
long_data['Avg_Spnd_Per_Clm'] = long_data['Avg_Spnd_Per_Clm'].astype(np.float64)  # DECIMAL(15,2)
long_data['Avg_Spnd_Per_Bene'] = long_data['Avg_Spnd_Per_Bene'].astype(np.float64)  # DECIMAL(15,2)
long_data['Outlier_Flag'] = long_data['Outlier_Flag'].astype(bool)  # BOOLEAN
```

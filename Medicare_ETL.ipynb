```python
import pandas as pd
import numpy as np
```

```python
# path for data
data_path = ("Medicare_Drugs.csv")

# reading file and creating data frame
raw_data = pd.read_csv(data_path)
raw_data = raw_data.fillna(0)
raw_data_df = pd.DataFrame(raw_data)

# formatting (need to fix and clean up!!!!!!)

pd.set_option('display.float_format', str)
raw_data_df.drop(raw_data_df.index[raw_data_df['Mftr_Name'] == 'Overall'], inplace = True)
raw_data_df.to_csv('raw_Medicare_data_transformed.csv', index=False)
raw_data_df.head(25)
```

```python
raw_data_df.dtypes
```

```python
# finding unique number of manufacturers
unique_mftr = raw_data_df.nunique()
unique_mftr
```

```python
# finding the how much was spent on each mftr
    # un comment "# total_manufact_dose_2018" to show df

total_mftr_spent_2018 = raw_data_df.groupby(["Mftr_Name"])["Tot_Spndng_2018"].sum()
# total_manufact_dose_2018
total_mftr_spent_2019 = raw_data_df.groupby(["Mftr_Name"])["Tot_Spndng_2019"].sum()
# total_total_manufact_spent_2019
total_mftr_spent_2020 = raw_data_df.groupby(["Mftr_Name"])["Tot_Spndng_2020"].sum()
# total_total_manufact_spent_2020
total_mftr_spent_2021 = raw_data_df.groupby(["Mftr_Name"])["Tot_Spndng_2021"].sum()
# total_total_manufact_spent_20221
total_mftr_spent_2022 = raw_data_df.groupby(["Mftr_Name"])["Tot_Spndng_2022"].sum()
# total_total_manufact_spent_2022

```

```python
total_mftr_units_2018 = raw_data_df.groupby(["Mftr_Name"])["Tot_Dsg_Unts_2018"].sum()
# total_total_manufact_spen_2018
total_mftr_units_2019 = raw_data_df.groupby(["Mftr_Name"])["Tot_Dsg_Unts_2019"].sum()
# total_total_manufact_spent_2019
total_mftr_units_2020 = raw_data_df.groupby(["Mftr_Name"])["Tot_Dsg_Unts_2020"].sum()
# total_total_manufact_spent_2020
total_mftr_units_2021 = raw_data_df.groupby(["Mftr_Name"])["Tot_Dsg_Unts_2021"].sum()
# total_total_manufact_spent_2021
total_mftr_units_2022 = raw_data_df.groupby(["Mftr_Name"])["Tot_Dsg_Unts_2022"].sum()
# total_total_manufact_spent_2022
```

```python
mftr_spent_summary = pd.DataFrame({"2018 Medicare Spent":total_mftr_spent_2018, 
                                   "2019 Medicare Spent":total_mftr_spent_2019, 
                                   "2020 Medicare Spent":total_mftr_spent_2020, 
                                   "2021 Medicare Spent":total_mftr_spent_2021, 
                                  "2022 Medicare Spent":total_mftr_spent_2022,})
mftr_spent_summary.head(5)

mftr_prod_summary = pd.DataFrame ({"2018 Medicare Units Produced":total_mftr_units_2018,
                                   "2019 Medicare Units Produced":total_mftr_units_2019,
                                   "2020 Medicare Units Produced":total_mftr_units_2020,
                                   "2021 Medicare Units Produced":total_mftr_units_2021,
                                   "2022 Medicare Units Produced":total_mftr_units_2022})

mftr_spent_summary.to_csv('Medicare_mftr_spent_summary.csv', index=False)
```

```python
mftr_prod_summary = pd.DataFrame ({"2018 Medicare Units Produced":total_mftr_units_2018,
                                   "2019 Medicare Units Produced":total_mftr_units_2019,
                                   "2020 Medicare Units Produced":total_mftr_units_2020,
                                   "2021 Medicare Units Produced":total_mftr_units_2021,
                                   "2022 Medicare Units Produced":total_mftr_units_2022})

mftr_prod_summary.head(5)

mftr_prod_summary.to_csv('Medicare_mftr_prod_summary.csv', index=False)
```

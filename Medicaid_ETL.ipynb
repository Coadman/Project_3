```python
import pandas as pd
import numpy as np
```

```python
# path for data
data_path = ("Medicaid_Drugs.csv")

# reading file and creating data frame
raw_data = pd.read_csv(data_path)
raw_data = raw_data.fillna(0)
raw_data_df = pd.DataFrame(raw_data)

# formatting
pd.set_option('display.float_format', str)


raw_data_df.head(25)
```

```python
# finding unique number of manufacturers
unique_mftr = raw_data_df.nunique()
unique_mftr
```

```python
#Manufacturer $$$ Spent

total_manufact_spent_2018 = raw_data_df.groupby(["Mftr_Name"])["Tot_Spndng_2018"].sum()
total_manufact_spent_2019 = raw_data_df.groupby(["Mftr_Name"])["Tot_Spndng_2019"].sum()
total_manufact_spent_2020 = raw_data_df.groupby(["Mftr_Name"])["Tot_Spndng_2020"].sum()
total_manufact_spent_2021 = raw_data_df.groupby(["Mftr_Name"])["Tot_Spndng_2021"].sum()
total_manufact_spent_2022 = raw_data_df.groupby(["Mftr_Name"])["Tot_Spndng_2022"].sum()

#Manufacturer Units Produced
total_manufact_dose_2018 = raw_data_df.groupby(["Mftr_Name"])["Tot_Dsg_Unts_2018"].sum()
total_manufact_dose_2019 = raw_data_df.groupby(["Mftr_Name"])["Tot_Dsg_Unts_2019"].sum()
total_manufact_dose_2020 = raw_data_df.groupby(["Mftr_Name"])["Tot_Dsg_Unts_2020"].sum()
total_manufact_dose_2021 = raw_data_df.groupby(["Mftr_Name"])["Tot_Dsg_Unts_2021"].sum()
total_manufact_dose_2022 = raw_data_df.groupby(["Mftr_Name"])["Tot_Dsg_Unts_2022"].sum()
```

```python
manu_spent_summary = pd.DataFrame({"2018 Medicaid Spend":total_manufact_spent_2018, "2018 Medicaid Units Produced":total_manufact_dose_2018,
                                   "2019 Medicaid Spend":total_manufact_spent_2019, "2019 Medicaid Units Produced":total_manufact_dose_2019,
                                   "2020 Medicaid Spend":total_manufact_spent_2020, "2020 Medicaid Units Produced":total_manufact_dose_2020,
                                   "2021 Medicaid Spend":total_manufact_spent_2021, "2021 Medicaid Units Produced":total_manufact_dose_2021,
                                  "2022 Medicaid Spend":total_manufact_spent_2022, "2022 Medicaid Units Produced":total_manufact_dose_2022,})
manu_spent_summary
```

```python
import pandas as pd
import numpy as np
```

```python
# path for data
data_path = ("Medicaid_Drugs.csv")

# reading file and creating data frame
raw_data = pd.read_csv(data_path)
raw_data = raw_data.fillna(0)
raw_data_df = pd.DataFrame(raw_data)

raw_data_df.drop(raw_data_df.index[raw_data_df['Mftr_Name'] == 'Overall'], inplace = True)

# formatting
pd.set_option('display.float_format', str)


raw_data_df.head(25)
```

```python
raw_data_df.dtypes
```

```python
# finding unique number of manufacturers
unique_mftr = raw_data_df.nunique()
unique_mftr
```

```python
#Manufacturer $$$ Spent

total_manufact_spent_2018 = raw_data_df.groupby(["Mftr_Name"])["Tot_Spndng_2018"].sum()
total_manufact_spent_2019 = raw_data_df.groupby(["Mftr_Name"])["Tot_Spndng_2019"].sum()
total_manufact_spent_2020 = raw_data_df.groupby(["Mftr_Name"])["Tot_Spndng_2020"].sum()
total_manufact_spent_2021 = raw_data_df.groupby(["Mftr_Name"])["Tot_Spndng_2021"].sum()
total_manufact_spent_2022 = raw_data_df.groupby(["Mftr_Name"])["Tot_Spndng_2022"].sum()

#Manufacturer Units Produced
total_manufact_dose_2018 = raw_data_df.groupby(["Mftr_Name"])["Tot_Dsg_Unts_2018"].sum()
total_manufact_dose_2019 = raw_data_df.groupby(["Mftr_Name"])["Tot_Dsg_Unts_2019"].sum()
total_manufact_dose_2020 = raw_data_df.groupby(["Mftr_Name"])["Tot_Dsg_Unts_2020"].sum()
total_manufact_dose_2021 = raw_data_df.groupby(["Mftr_Name"])["Tot_Dsg_Unts_2021"].sum()
total_manufact_dose_2022 = raw_data_df.groupby(["Mftr_Name"])["Tot_Dsg_Unts_2022"].sum()
```

```python
#DF Creation

manu_spent_summary = pd.DataFrame({"2018 Medicaid Spend":total_manufact_spent_2018, "2018 Medicaid Units Produced":total_manufact_dose_2018,
                                   "2019 Medicaid Spend":total_manufact_spent_2019, "2019 Medicaid Units Produced":total_manufact_dose_2019,
                                   "2020 Medicaid Spend":total_manufact_spent_2020, "2020 Medicaid Units Produced":total_manufact_dose_2020,
                                   "2021 Medicaid Spend":total_manufact_spent_2021, "2021 Medicaid Units Produced":total_manufact_dose_2021,
                                   "2022 Medicaid Spend":total_manufact_spent_2022, "2022 Medicaid Units Produced":total_manufact_dose_2022,})




#formatting
manu_spent_summary["2018 Medicaid Spend"] = manu_spent_summary["2018 Medicaid Spend"].map("${:,.2f}".format)
manu_spent_summary["2019 Medicaid Spend"] = manu_spent_summary["2019 Medicaid Spend"].map("${:,.2f}".format)
manu_spent_summary["2020 Medicaid Spend"] = manu_spent_summary["2020 Medicaid Spend"].map("${:,.2f}".format)
manu_spent_summary["2021 Medicaid Spend"] = manu_spent_summary["2021 Medicaid Spend"].map("${:,.2f}".format)
manu_spent_summary["2022 Medicaid Spend"] = manu_spent_summary["2022 Medicaid Spend"].map("${:,.2f}".format)

manu_spent_summary_rounded = manu_spent_summary.round(0)
manu_spent_summary_rounded
```

```python
#total manufactuer of drugs

total_drug_manufact = raw_data_df.groupby(["Brnd_Name"])["Tot_Mftr"].sum()

#total spent on drug

total_drug_spend_2018 = raw_data_df.groupby(["Brnd_Name"])["Tot_Spndng_2018"].sum()
total_drug_spend_2019 = raw_data_df.groupby(["Brnd_Name"])["Tot_Spndng_2019"].sum()
total_drug_spend_2020 = raw_data_df.groupby(["Brnd_Name"])["Tot_Spndng_2020"].sum()
total_drug_spend_2021 = raw_data_df.groupby(["Brnd_Name"])["Tot_Spndng_2021"].sum()
total_drug_spend_2022 = raw_data_df.groupby(["Brnd_Name"])["Tot_Spndng_2022"].sum()

drug_spent_summary = pd.DataFrame({"Total # of Manufacturers":total_drug_manufact, "2018 Spent on Drug":total_drug_spend_2018,
                                    "2019 Spent on Drug":total_drug_spend_2019,
                                   "2020 Spent on Drug":total_drug_spend_2020,
                                   "2021 Spent on Drug":total_drug_spend_2021,
                                  "2022 Spent on Drug":total_drug_spend_2022,})


#more formatting
drug_spent_summary["2018 Spent on Drug"] = drug_spent_summary["2018 Spent on Drug"].map("${:,.2f}".format)
drug_spent_summary["2019 Spent on Drug"] = drug_spent_summary["2019 Spent on Drug"].map("${:,.2f}".format)
drug_spent_summary["2020 Spent on Drug"] = drug_spent_summary["2020 Spent on Drug"].map("${:,.2f}".format)
drug_spent_summary["2021 Spent on Drug"] = drug_spent_summary["2021 Spent on Drug"].map("${:,.2f}".format)
drug_spent_summary["2022 Spent on Drug"] = drug_spent_summary["2022 Spent on Drug"].map("${:,.2f}".format)


#sorting by manufacturer
total_drug_manufact_sorted = drug_spent_summary.sort_values(by="Total # of Manufacturers", ascending=False)

total_drug_manufact_sorted.head(25)
```

```python
total_drug_manufact_sorted.to_csv("medicaid_drug_brands_transformed.csv", encoding='utf8', index=False)
manu_spent_summary_rounded.to_csv("medicaid_manufacturers_transformed.csv", encoding='utf8', index=False)
raw_data_df.to_csv("raw_data_transformed.csv", encoding='utf8', index=False)
```
